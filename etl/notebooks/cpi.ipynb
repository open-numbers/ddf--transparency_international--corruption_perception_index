```python
all_data = {}

for f in os.listdir('source/'):
    if 'xls' in f:
        if '2014' in f:
            data = pd.read_excel(os.path.join('source', f), skiprows=1)
            data = data[['Country / Territory', 'Unnamed: 2', 'CPI 2014 Score']]  ## unnamed 2 is wbcode
            data['year'] = 2014
            all_data['2014'] = data
        if '2015' in f:
            data = pd.read_excel(os.path.join('source', f))
            data = data[['Country', 'wbcode', 'CPI2015']]
            data['year'] = 2015
            all_data['2015'] = data
        if '2010' in f:
            data = pd.read_excel(os.path.join('source', f), skiprows=1)
            data = data[['Country / Territory', 'CPI 2010 Score']]
            data['year'] = 2010
            all_data['2010'] = data
        if '2011' in f:
            data = pd.read_excel(os.path.join('source', f))
            data = data[['country', 'cpi11']]
            data['year'] = 2011
            all_data['2011'] = data
        if '2012' in f:
            data = pd.read_excel(os.path.join('source', f))
            data = data[['Country / Territory', 'CPI 2012 Score']]
            data['year'] = 2012
            all_data['2012'] = data
        if '2013' in f:
            data = pd.read_excel(os.path.join('source', f))
            data = data[['Country / Territory', 'Unnamed: 2', 'CPI 2013 Score']]  ## unnamed 2 is wbcode
            data['year'] = 2013
            all_data['2013'] = data
```

```python
all_data['2014'] = all_data['2014'].dropna()

all_data['2014'].columns = ['country', 'wbcode', 'cpi', 'year']
```

```python
all_data['2013'] = all_data['2013'].dropna()

all_data['2013'].columns = ['country', 'wbcode', 'cpi', 'year']
```

```python
all_data['2012'] = all_data['2012'].dropna()

all_data['2012'].columns = ['country', 'cpi', 'year']
```

```python
all_data['2011'] = all_data['2011'].dropna()

all_data['2011'].columns = ['country', 'cpi', 'year']
```

```python
all_data['2010'] = all_data['2010'].drop([0, 1]).dropna()

all_data['2010'].columns = ['country', 'cpi', 'year']
```

```python
all_data['2015'] = all_data['2015'].dropna()

all_data['2015'].columns = ['country', 'wbcode', 'cpi', 'year']
```

```python
all_data_df = pd.concat(all_data.values())
```

```python
all_data_df = all_data_df.reset_index(drop=True)
```

```python
gps = all_data_df.groupby('country')
```

```python
for k, v in gps.groups.items():
    
    df = all_data_df[all_data_df['country'] == k]
    
    wbcode_list = df['wbcode'].unique()
    wbcode_list = [x for x in wbcode_list if x is not np.nan]
    
    if len(wbcode_list) == 0:
        country = to_concept_id(k)
        
    else:
        country = wbcode_list[0]
        
    all_data_df.loc[v, 'wbcode'] = country
```
