```python
def get_data(fn, country_column, score_column, sheet_name=0, **kwargs):
    if fn.endswith('xls') or fn.endswith('xlsx'):
        data = pd.read_excel(fn, sheet_name, **kwargs)
    else:
        data = pd.read_csv(fn, **kwargs)
        
    df = data[[country_column, score_column]].copy()
    df.columns = ['country', 'cpi']
    df['country'] = df['country'].str.strip()
    
    return df.copy()
```
